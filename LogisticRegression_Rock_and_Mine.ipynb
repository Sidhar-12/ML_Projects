```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv('/content/SVM techniques.csv',header=None)
#print(sonar_data)
sonar_data.head() #This function shons first 5 data set of our data set
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[4].value_counts()
```

```python
x=sonar_data.drop(columns=4,axis=1)
y=sonar_data[4]
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,stratify=y,random_state=2)
#print(x_train)
#print(y_train)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_predict=model.predict(x_train)
x_train_accuracy=accuracy_score(x_train_predict,y_train)
print(x_train_accuracy)
```

```python

x_test_accuracy=model.score(x_test,y_test)
print(x_test_accuracy)
```
